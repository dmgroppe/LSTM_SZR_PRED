```python
def norm_eeg(eeg):
    md=np.median(eeg)
    iqr=stats.iqr(eeg)
    print('Median={}, IQR={}'.format(md,iqr))
    shrinkage=3*iqr
    eeg=(eeg-md)/(shrinkage)
    return eeg, md, shrinkage

def rand_eeg_samp(eeg,n):
    bro=eeg.flatten()
    return bro[np.random.randint(0,len(bro),n)]
```

```python
raw=np.squeeze(mat['raw_eeg'][0,:,:])
raw, raw_md, raw_shrinkage=norm_eeg(raw);
clean=np.squeeze(mat['cleaned_eeg'][0,:,:])
clean, clean_md, clean_shrinkage=norm_eeg(clean);
blink=np.squeeze(mat['blink_eeg'][0,:,:])
blink, blink_md, blink_shrinkage=norm_eeg(blink);
```

```python
plt.figure(1)
plt.clf()
plt.hist(rand_eeg_samp(raw,10000),100);
plt.title('Raw')

plt.figure(2)
plt.clf()
plt.hist(rand_eeg_samp(clean,10000),100);
plt.title('Clean')

plt.figure(3)
plt.clf()
plt.hist(rand_eeg_samp(blink,10000),100);
plt.title('Blink')
```
